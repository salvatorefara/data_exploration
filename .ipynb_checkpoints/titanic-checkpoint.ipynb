# Titanic dataset

### Attributes
1. survived - Survival (0 = No; 1 = Yes)
2. pclass - Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)
3. name - Name
4. sex - Sex
5. age - Age
6. sibsp - Number of Siblings/Spouses Aboard
7. parch - Number of Parents/Children Aboard
8. ticket - Ticket Number
9. fare - Passenger Fare
10. cabin - Cabin
11. embarked - Port of Embarkation (C = Cherbourg; Q = Queenstown; S = 12. Southampton)
12. boat - Lifeboat (if survived)
13. body - Body number (if did not survive and body was recovered)
14. home.dest - Hometown

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import sklearn as sk

# utility function to convert to string to float, which handles missing 
# values as NaNs
nan_float = lambda data: np.array([np.nan if x=='?' else float(x) for x in data])
```

```python
# load dataset
data_path = 'https://www.openml.org/data/get_csv/16826755/phpMYEkMl'
data = pd.read_csv(data_path)

data_attrib = data.keys().to_list()
print(data_attrib)
```

```python
# extract relevant features from attributes
features = {
    'sex': 1*(data['sex'].to_numpy()=='female'),
    'class': data['pclass'].to_numpy(),
    'age': nan_float(data['age']),
    'siblings/spouses': data['sibsp'].to_numpy(),
    'parents/children': data['parch'].to_numpy(),
    'fare': nan_float(data['fare'])
    }

# labels
Y = data['survived'].to_numpy()

# concatenated features
X = np.concatenate([x[:,np.newaxis] for x in features.values()], axis=1)
n = np.shape(X)[1]
X_names = [x for x in features.keys()]
```

```python
# show feature histograms
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(20,10))
for k,item in enumerate(features.items()):
    plt.subplot(2,3,k+1)
    plt.hist(item[1])
    plt.xlabel(item[0])
    plt.ylabel('count')    
```

```python
# gender distribution across classes
plt.figure()
plt.hist([features['class'][features['sex']==0],features['class'][features['sex']==1]])
plt.xlabel('class')
plt.legend(['male','female']);
```

```python
# show feature boxplots using 'class' as grouping variable
group = 'class'

plt.figure(figsize=(20,10))

curr_feat = 'age'
plt.subplot(221)
x = features[curr_feat]
y = features[group][~np.isnan(x)]
x = x[~np.isnan(x)]
plt.boxplot([x[y==1],x[y==2],x[y==3]])
plt.xlabel(group)
plt.ylabel(curr_feat);

curr_feat = 'fare'
plt.subplot(222)
x = features[curr_feat]
y = features[group][~np.isnan(x)]
x = x[~np.isnan(x)]
plt.boxplot([x[y==1],x[y==2],x[y==3]])
plt.xlabel(group)
plt.ylabel(curr_feat);

curr_feat = 'siblings/spouses'
plt.subplot(223)
x = features[curr_feat]
y = features[group][~np.isnan(x)]
x = x[~np.isnan(x)]
plt.boxplot([x[y==1],x[y==2],x[y==3]])
plt.xlabel(group)
plt.ylabel(curr_feat);

curr_feat = 'parents/children'
plt.subplot(224)
x = features[curr_feat]
y = features[group][~np.isnan(x)]
x = x[~np.isnan(x)]
plt.boxplot([x[y==1],x[y==2],x[y==3]])
plt.xlabel(group)
plt.ylabel(curr_feat);
```

```python
plt.figure(figsize=(8,3))
plt.bar(np.arange(n), np.sum(np.isnan(X),axis=0))
plt.xticks(ticks=np.arange(n), labels=X_names, rotation=45)
plt.ylabel('# missing values');
```
