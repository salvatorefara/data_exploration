# Heart disease dataset

In this notebook, I analyse the heart disease dataset published in the [UCI repository](https://archive.ics.uci.edu/ml/datasets/heart+disease). As described in the repository, this dataset contains 14 attributes. The standard task with this dataset is to predict the presence of coronary heart disease, which is encoded by the attribute 'num'. 

Here, I perform a quick analysis of the dataset in order to find out what factors are associated with the incidence of heart disease.

### Attribute description

Description of the dataset attributes (as provided in the [UCI repository](https://archive.ics.uci.edu/ml/datasets/heart+disease)):

1. age: age in years
2. sex: sex (1 = male; 0 = female)
3. cp: chest pain type
        -- Value 1: typical angina
        -- Value 2: atypical angina
        -- Value 3: non-anginal pain
        -- Value 4: asymptomatic
4. trestbps: resting blood pressure (in mm Hg on admission to the hospital) 
5. chol: serum cholestoral in mg/dl 
6. fbs: (fasting blood sugar > 120 mg/dl)  (1 = true; 0 = false) 
7. restecg: resting electrocardiographic results
        -- Value 0: normal
        -- Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
        -- Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria 
8. thalach: maximum heart rate achieved 
9. exang: exercise induced angina (1 = yes; 0 = no) 
10. oldpeak: ST depression induced by exercise relative to rest 
11. slope: the slope of the peak exercise ST segment
        -- Value 1: upsloping
        -- Value 2: flat
        -- Value 3: downsloping
12. ca: number of major vessels (0-3) colored by fluoroscopy 
13. thal: $\beta$-Thalassemia
        -- Value 3: normal
        -- Value 6: fixed defect
        -- Value 7: reversable defect
14. num: diagnosis of heart disease (angiographic disease status)
        -- Value 0: < 50% diameter narrowing
        -- Value 1,2,3,4: > 50% diameter narrowing

## 1 - Data processing

### Download the dataset

Run the next cell to download the dataset.

```python
# manually specify the attributes (extracted from the dataset description),
# since the .csv file doesn't have any
attributes = np.array(['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num'])

# from the attribute description, we see that the following are categorical
cat_attributes = np.array(['sex','cp','restecg','exang','thal'])
noncat_attributes = attributes[~ismember(attributes,cat_attributes)]

# load dataset
data_path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'
data = pd.read_csv(data_path, header=None, names=attributes)

display(data.head())
display(data.info())
```

### Missing values

Check if the dataset has some missing values (encoded by '?').

```python
# show missing values for each attribute (encoded by '?')
missing_idx = data=='?'
print('\n# of missing values:')
print(missing_idx.sum(axis=0))
```

We see that there are some missing values in the 'ca' and 'thal' attributes. Run the next cell to reject the examples (rows in the pandas dataframe) containing missing values.  

```python
# reject rows with missing values (for now)
data = data[~missing_idx.any(axis=1)]

print(data.shape)
```

### Categorical and non-categorical attributes

The dataset contains both categorical and non-categorical attributes. Run the next cell to perform the following operations:
- ensure that the values of categorical attributes are strings with integer formatting (e.g. '1','2','3',...) 
- ensure that the values of non-categorical attributes are floats
- threshold the values of attribute 'num'. This attribute has values 0 (no disease) and 1,2,3,4 (various levels of disease). Here we convert all non-zero values into 1.

```python
# convert values of categorical attributes to strings
for attr in cat_attributes:
    unique_val = data[attr].unique()
    for val in unique_val:
        data.loc[data[attr]==val, attr] = str(int(float(val)))

# convert values of non-categorical attributes to float
for attr in noncat_attributes:
    unique_val = data[attr].unique()
    for val in unique_val:
        data.loc[data[attr]==val, attr] = float(val)
    data[attr] = pd.to_numeric(data[attr])
        
# show unique values for the categorical attributes
print('Unique values for categorical attributes: ')
for attr in cat_attributes:
    print(attr + ': ' + str(data[attr].unique()))

# threshold 'num' (0 = absence of disease, 1 = presence of disease) 
data['num'] = 1.*(data['num']>=1)

# show unique values of 'num'
attr = 'num'
print('\nUnique values of attribute to predict: ')
print(attr + ': ' + str(data[attr].unique()))
    
display(data.head())
display(data.info())
```

### Standardization

The next cell standardizes all non-categorical attributes (except 'num').

```python
# non-categorical features = non-categorical attributes except 'num'
noncat_features = noncat_attributes[~ismember(noncat_attributes,'num')]

# standardize non-categorical fetures
standardize = lambda x: (x - x.mean())/x.std()
data[noncat_features] = data[noncat_features].apply(standardize)

# show results
print('Mean (non-categorical attributes):')
print(data[noncat_attributes].mean())
print('\nStandard deviation (non-categorical attributes): ')
print(data[noncat_attributes].std())
```

### One-hot encoding

The next cell applies one-hot encoding to the non-binary categorical features. It could in principle be applied to binary categorical features too, but it would be redundant, since binary categorical features can be fully encoded by a single column of 0s and 1s.

```python
# use one-hot encoding for all non-binary categorical attributes
nonbin_cat_attributes = \
    cat_attributes[[len(data[attr].unique())>2 for attr in cat_attributes]]
data_onehot = pd.get_dummies(data, columns=nonbin_cat_attributes, prefix_sep='')

display(data_onehot.head())
display(data_onehot.info())
```

## 2 - Statistical modelling

In the next cell, we fit a logistic model to predict 'num' from all other attributes. 

```python
# specify the model
attributes_all = data_onehot.columns.to_numpy()
var_to_pred = 'num'
features = attributes_all[~ismember(attributes_all, var_to_pred)]
model_specs = var_to_pred + ' ~ ' + ' + '.join(features.tolist())

# fit logistic regression model(s)
model = sm.GLM.from_formula(model_specs, family=sm.families.Binomial(), data=data_onehot)
result = model.fit()
```

## 3 - Results

Let's visualize the results of the logistic regression:

```python
display(result.summary())

plt.figure(figsize=(3.25,7))
av = result.params[1:]
err = result.conf_int()[1][1:] - av
pos = np.flip(np.arange(len(av)))
plt.errorbar(x=av, y=pos, xerr=err, fmt=' ', color='b')
plt.scatter(av, pos)
plt.yticks(ticks=pos, labels=result.params.index[1:])
plt.vlines(0,*plt.ylim(), color='k')
plt.autoscale(enable=True, axis='y', tight=True)
plt.xlabel('coeff. value +- 95% CI');
```

As we see above, the logistic regression revealed a significant association between the probability of heart disease and the following factors:
- **sex**: the positive coefficient indicates that the probability of heart disease is higher in men compared to women
- **trestbps**: the probability of disease increases with resting blood pressure, although the strength of this association is relatively weak
- **ca**: the probability of disease increases with the number of major vessels colored by fluoroscopy
- **cp1, cp3, cp4**: chest pain type 1 and 3 are associated with lower probability of disease, whereas chest pain type 4 is associated with higher probability of disease. This result is puzzling since according to the attribute description in the UCI depository, chest pain type 4 stands for 'asymptomatic'. So, no chest pain = higher probability of disease
- **thal3, thal7**: types of $\beta$-Thalassemia are also associated with heart disease: 'normal' (thal3) is associated with lower probability, whereas 'reversible defect' (thal7) is associated with higher probability. 

The next cell plots these significant factors relative to the disease groups.

```python
group = 'num'
group_labels = ['no disease', 'disease']
group_values = data[group].unique()
feat_vec = ['sex','trestbps','ca','cp','thal']
plot_type = ['bar','box','box','bar','bar']
nfeat = len(feat_vec)
nrows = int(np.ceil(nfeat/2))
ncols = 2

plt.figure(figsize=(10,3.25*nrows))
for k,feat in enumerate(feat_vec):
    if plot_type[k]=='bar':
        feat_values = data[feat].unique()
        bottom = 0
        plt.subplot(nrows,ncols,k+1)
        for val in feat_values:
            labels, counts = np.unique(data.loc[data[feat] == val, group], return_counts=True)
            plt.bar(labels, counts, align='center', bottom = bottom, width=bar_width)
            bottom += counts
        plt.xticks(ticks=labels, labels=group_labels)
        plt.ylabel('people count')
        plt.xlim(xlimits_bar)
        plt.legend([feat + ': ' + str(val) for val in feat_values]);
    elif plot_type[k]=='box':
        plt.subplot(nrows,ncols,k+1)
        plt.boxplot([data.loc[data[group]==val, feat] for val in group_values], labels=group_labels)
        plt.xlim(xlimits_box)
        plt.ylabel(feat);

plt.tight_layout();
```
